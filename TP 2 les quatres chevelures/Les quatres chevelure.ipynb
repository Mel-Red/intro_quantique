```python
from qiskit import *
```

```python
# On crée le circuit quantique
```

```python
q = QuantumRegister(8)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)
```

```python
qc.draw('mpl')
```

```python
# Les participants sont associés à un Qbit Elodie qui va déterminer
# en premier la parité est q0 les autres sont q1 q2 q3
# Le résultat d'Elodie sur la parité est stocké dans q4
# Les résultats des autres participants sont stockés dans les qbits
# suivants: q5 q6 q7
```

```python
# On place une porte hadamar sur les Qbits généré
# Afin de permettre la superpositions des solutions de l'énigme
```

```python
qc.h(0)
qc.h(1)
qc.h(2)
qc.h(3)
```

```python
qc.draw('mpl')
```

```python
# On applique les portes CNOT afin de transcrire ce que voi Elodie
```

```python
qc.cx(1,4)
qc.cx(2,4)
qc.cx(3,4)
```

```python
qc.draw('mpl')
```

```python
# Etape suivante
```

```python
qc.barrier()
```

```python
# Les autres participant récupère le résultat dans le qbit 4 
```

```python
qc.cx(4,5)
qc.cx(4,6)
qc.cx(4,7)
```

```python
qc.draw('mpl')
```

```python
# Etape suivante
```

```python
# Les participants vont tour à tour déterminer 
# la couleur de leur chevelure et vont l'annoncer puis
# les stocker dans leur qbits de résultats
```

```python
qc.barrier()
```

```python
# Premier participant après Elodie
# Il détermine les couleurs des qbits q2 et q3
# On utilise les portes CNOT pour mettre les résultats dans
# le qbits de résultats q5
```

```python
qc.cx(2,5)
qc.cx(3,5)
```

```python
qc.draw('mpl')
```

```python
# Les participants restant récupèrent le résultats
# à l'aide de portes CNOT
```

```python
qc.cx(5,6)
qc.cx(5,7)
```

```python
qc.draw('mpl')
```

```python
# Le participant 3 détermine la couleur du qbit q3
# et le note dans son qbits de résultat
```

```python
qc.barrier()
```

```python
qc.cx(3,6)
```

```python
qc.draw('mpl')
```

```python
qc.barrier()
```

```python
# Le dernier participant récupère l'information sur la couleur 
# de ses cheveux et la note dans son qbit de résultat q7
```

```python
qc.cx(6,7)
```

```python
qc.draw('mpl')
```

```python
# On prend ensuite la mesure des qbits du cicruit
```

```python
qc.measure(0,0)
qc.measure(1,0)
qc.measure(2,0)
qc.measure(3,0)
qc.measure(4,0)
qc.measure(5,0)
qc.measure(6,0)
qc.measure(7,0)
```

```python
qc.draw('mpl')
```
