```python
from qiskit import *
```

```python
# On cr√©e le circuit quantique
```

```python
q = QuantumRegister(8)
c = ClassicalRegister(1)
qc = QuantumCircuit(q,c)
```

```python
qc.draw('mpl')
```

```python
# Les participants sont associ√©s √† un Qbit Elodie qui va d√©terminer
# en premier la parit√© est q0 les autres sont q1 q2 q3
# Le r√©sultat d'Elodie sur la parit√© est stock√© dans q4
# Les r√©sultats des autres participants sont stock√©s dans les qbits
# suivants: q5 q6 q7
```

```python
# On place une porte hadamar sur les Qbits g√©n√©r√©
# Afin de permettre la superpositions des solutions de l'√©nigme
```

```python
qc.h(0)
qc.h(1)
qc.h(2)
qc.h(3)
```

```python
qc.draw('mpl')
```

```python
# On applique les portes CNOT afin de transcrire ce que voi Elodie
```

```python
qc.cx(1,4)
qc.cx(2,4)
qc.cx(3,4)
```

```python
qc.draw('mpl')
```

```python
# Etape suivante
```

```python
qc.barrier()
```

```python
# Les autres participant r√©cup√®re le r√©sultat dans le qbit 4 
```

```python
qc.cx(4,5)
qc.cx(4,6)
qc.cx(4,7)
```

```python
qc.draw('mpl')
```

```python
# Etape suivante
```

```python
# Les participants vont tour √† tour d√©terminer 
# la couleur de leur chevelure et vont l'annoncer puis
# les stocker dans leur qbits de r√©sultats
```

```python
qc.barrier()
```

```python
# Premier participant apr√®s Elodie
# Il d√©termine les couleurs des qbits q2 et q3
# On utilise les portes CNOT pour mettre les r√©sultats dans
# le qbits de r√©sultats q5
```

```python
qc.cx(2,5)
qc.cx(3,5)
```

```python
qc.draw('mpl')
```

```python
# Les participants restant r√©cup√®rent le r√©sultats
# √† l'aide de portes CNOT
```

```python
qc.cx(5,6)
qc.cx(5,7)
```

```python
qc.draw('mpl')
```

```python
# Le participant 3 d√©termine la couleur du qbit q3
# et le note dans son qbits de r√©sultat
```

```python
qc.barrier()
```

```python
qc.cx(3,6)
```

```python
qc.draw('mpl')
```

```python
qc.barrier()
```

```python
# Le dernier participant r√©cup√®re l'information sur la couleur 
# de ses cheveux et la note dans son qbit de r√©sultat q7
```

```python
qc.cx(6,7)
```

```python
qc.draw('mpl')
```

```python
# On prend ensuite la mesure des qbits du cicruit
```

```python
qc.measure(0,0)
qc.measure(1,0)
qc.measure(2,0)
qc.measure(3,0)
qc.measure(4,0)
qc.measure(5,0)
qc.measure(6,0)
qc.measure(7,0)
```
