```python
from qiskit import *
```

```python
# On créer un registre quantique de 3 bits
```

```python
q = QuantumRegister(3)
```

```python
# On créer un registre
```

```python
c = ClassicalRegister(1)
```

```python
# On utilise C et Q pour créer un circuit quantique
```

```python
circ = QuantumCircuit(q,c)
```

```python
circ.draw('mpl')
```

```python
# Portes hadamar
```

```python
circ.h(0)
```

```python
circ.draw('mpl')
```

```python
circ.h(2)
```

```python
circ.draw('mpl')
```

```python
 # CNOT
```

```python
circ.cx(0,1)
```

```python
circ.draw('mpl')
```

```python
circ.barrier(0,1,2)
```

```python
circ.draw('mpl')
```

```python
circ.cx(2,1)
```

```python
circ.draw('mpl')
```

```python
# On restaure l'état initial
```

```python
circ.x(2)
```

```python
circ.draw('mpl')
```

```python
circ.cx(2,0)
```

```python
circ.draw('mpl')
```

```python
circ.x(2)
```

```python
# Pour le second état, on fait une porte de permutation.
```

```python
circ.swap(0,1)
```

```python
circ.draw('mpl')
```

```python
circ.x(0)
```

```python
circ.x(1)
```

```python
circ.draw('mpl')
```

```python
circ.cx(2,1)
```

```python
circ.x(2)
```

```python
circ.draw('mpl')
```

```python
circ.cx(2,0)
```

```python
circ.x(2)
```

```python
circ.draw('mpl')
```

```python
circ.draw('mpl')
```
